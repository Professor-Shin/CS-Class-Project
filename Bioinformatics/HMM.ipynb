### Algorithm

```python
def data_preparation(state,base,prob_state,transition,emission,sequence): # for Algorithm
  states = [s.strip() for s in state]  # Remove newline characters from state
  observations = sequence

  start_probability = {states[i]: float(prob_state[i].strip()) for i in range(len(states))}  # Create start_probability dictionary

  transition_probability = {}
  for i, s in enumerate(states):
    transition_probability[s] = {}
    trans_probs = transition[i].strip().split()
    for j, s2 in enumerate(states):
        transition_probability[s][s2] = float(trans_probs[j])  # Create transition_probability dictionary

  emission_probability = {}
  for i, s in enumerate(states):
    emission_probability[s] = {}
    emit_probs = emission[i].strip().split()
    for j, b in enumerate(base):
        emission_probability[s][b.strip()] = float(emit_probs[j])  # Create emission_probability dictionary

  return states, observations, start_probability, transition_probability, emission_probability
state,sequence,prob_state,transition,emission = data_preparation(state,base,prob_state,transition,emission,sequence)
base = [b.strip() for b in base]
```

```python
def viterbi(obs, states, start_p, trans_p, emit_p):
    V = [{}]
    path = {}

    # Initialize base cases (t == 0)
    for y in states:
        V[0][y] = start_p[y] * emit_p[y][obs[0]]
        path[y] = [y]

    # Run Viterbi for t > 0
    for t in range(1, len(obs)):
        V.append({})
        newpath = {}

        for y in states:
            (prob, state) = max((V[t-1][y0] * trans_p[y0][y] * emit_p[y][obs[t]], y0) for y0 in states)
            V[t][y] = prob
            newpath[y] = path[state] + [y]

        # Don't need to remember the old paths
        path = newpath

    n = 0           # if only one element is observed max probability is at the first state
    if len(obs) != 1:
      n = t
    (prob, state) = max((V[n][y], y) for y in states)
    return (prob, path[state])

# Example usage (replace with your actual data)
states = state
observations = sequence
start_probability = prob_state
transition_probability = transition
emission_probability = emission
prob, path = viterbi(observations, states, start_probability, transition_probability, emission_probability)
print("Most likely sequence of hidden states:", path)
print("Probability of this sequence:", prob)
```
